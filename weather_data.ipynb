# **Weather Dataset**


# ***Columns***

| **Column Name**    | **Description**                                                                                                       | **Meaning**                                                |
| ------------------ | --------------------------------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------- |
| `Date/Time`        | The **timestamp** of the weather reading. Format: `MM/DD/YYYY HH:MM` (24-hour).                                       | The exact date and time when weather data was recorded.                    |
| `Temp_C`           | **Air temperature** in degrees Celsius (`°C`).                                                                        | The hotness or coldness of air at that time.                               |
| `Dew Point Temp_C` | **Dew point** temperature in °C. Indicates the temperature at which air is saturated with moisture.                   | The temperature at which water drops (dew) start to form in the air.       |
| `Rel Hum_%`        | **Relative Humidity** in percentage (`%`). Shows how much moisture is in the air compared to the maximum it can hold. | How full the air is with water; 100% means it can't hold more.             |
| `Wind Speed_km/h`  | **Wind speed** in kilometers per hour (`km/h`).                                                                       | How fast the air (wind) is moving.                                         |
| `Visibility_km`    | **Visibility distance** in kilometers. Shows how far one can see clearly.                                             | How far a person can see things in front of them.                          |
| `Press_kPa`        | **Atmospheric pressure** in kilopascals (`kPa`). Lower values can indicate storms or bad weather.                     | The weight of the air pressing down; low values often mean rain or storms. |
| `Weather`          | **Weather condition** description, like *Fog*, *Freezing Drizzle*, *Mostly Cloudy*, etc.                              | Words that describe what's happening in the sky, like fog or rain.         |

```python
hourly_temp = df1.groupby('Time_H')['Temp_C'].mean()
hourly_temp
```

#  ***Visualize the Trends***

```python
plt.figure(figsize=(10,7))
plt.plot(hourly_temp.index,hourly_temp.values,marker='o')
plt.title('Average Temperature By Hours of Day.')
plt.xlabel('Hour of Day')
plt.ylabel('Average Temperature')
plt.grid()
plt.xticks(range(0,24))
plt.show()
```

# ***Analyze Weather Patterns***	
- When was the fog most common?
- What kind of weather is linked with low visibility or high humidity?

```python
df1.head(2)
```

```python
fog_df = df1[df1['Weather'].str.contains('Fog',na=False)]
fog_df
```

```python
fog_df1 = df1[df1['Weather'] == 'Fog']
fog_df1
```

```python
# .size() counts how many times fog happened on each day.
fog_counts = fog_df1.groupby(['Year','Month','Day']).size().reset_index(name='Time_H')
fog_counts
```

```python
max_fog_day = fog_counts.sort_values(by='Time_H', ascending=False).iloc[0]
print(f'Fog Occurs The Most on {max_fog_day['Year']}-{max_fog_day['Month']}-{max_fog_day['Day']} with {max_fog_day['Time_H']} hours')
```

- What kind of weather is linked with low visibility or high humidity?

```python
df1.head(2)
```

```python
low_visibility_df1 = df1[df1['Visibility_km'] < 2.0]
low_visibility_df1
```

```python
plt.figure(figsize=(10,8))
weather_low_visi=low_visibility_df1['Weather'].value_counts()
plt.pie(weather_low_visi,
        labels=weather_low_visi.index,
        autopct='%2.1f%%',
       
        shadow=True,
        colors = sns.color_palette('pastel')[0:4],
        textprops={'fontsize':12}
        )
plt.axis('equal')
plt.title('Weather Contains Lower Visibility.',fontsize=14)
plt.show()
```

- Find out which month had most fog or rain

```python
fog_df = df1[df1['Weather'].str.contains('Fog', case=False, na=False)]
most_foggy_month = fog_df.groupby('Month').size().idxmax()
print("Month with most fog:", most_foggy_month)


```

 # ***case= False***
  This ignores case sensitivity.

 So it matches "fog", "Fog", "FOG", etc.

# ***na=False***
This handles missing values (i.e., NaN in 'Weather' column).

Without this, if any row has NaN, it would raise an error.

na=False means: treat missing values as if they do not match (i.e., False).

```python
fog_df = df1[df1['Weather'].str.contains('Rain', case=False, na=False)]
most_rainy_month = fog_df.groupby('Month').size().idxmax()
print("Month with most rain:", most_rainy_month)


```
